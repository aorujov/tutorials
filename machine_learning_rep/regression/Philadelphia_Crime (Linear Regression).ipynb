# Philadelphia Crime Rate

### Model: Linear Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('Philadelphia_Crime_Rate_noNA.csv')
```

### Model 1

Predict house prices using the crime rate

```python
var = 'CrimeRate'
df.plot.scatter(x=var, y='HousePrice')

model1 = LinearRegression()
model1.fit(df['CrimeRate'].to_frame(), df['HousePrice'])
```

```python
#plot
plt.plot(df['CrimeRate'], df['HousePrice'], '.',
         df['CrimeRate'], model1.predict(df['CrimeRate'].to_frame()), '-')
```

### Model 2

Predict house prices using the crime rate after removing the outlier in the above plot

```python
# remove outlier
df = df[df['MilesPhila'] != 0.0]
# rescatter
df.plot.scatter(x=var, y='HousePrice')
```

```python
model2 = LinearRegression()
model2.fit(df['CrimeRate'].to_frame(), df['HousePrice'])

#plot
plt.plot(df['CrimeRate'], df['HousePrice'], '.',
         df['CrimeRate'], model2.predict(df['CrimeRate'].to_frame()), '-')
```

#### Compare two models

```python
# model comparison
print("model 1: {}  {}".format(model1.intercept_, model1.coef_))
print("model 2: {}  {}".format(model2.intercept_, model2.coef_))
```

### Model 3

Fit the model after removing houses with price > 350k

```python
df = df[df['HousePrice'] < 350000]
model3 = LinearRegression()
model3.fit(df['CrimeRate'].to_frame(), df['HousePrice'])


#plot
plt.plot(df['CrimeRate'], df['HousePrice'], '.',
         df['CrimeRate'], model2.predict(df['CrimeRate'].to_frame()), '-')


# model coefficients
print("model 3: {}  {}".format(model3.intercept_, model3.coef_))
```
